```python
# Import required libraries
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML
import matplotlib.pyplot as plt
from collections import defaultdict
from datetime import datetime

# Import the LMS system module
from lms_system import (
    AIAssistant,
    initialize_sample_data,
    render_teacher_dashboard,
    render_student_dashboard,
    render_login_screen
)

print("âœ… All libraries loaded successfully!")
print("ğŸ“š K-12 Learning Management System - Ready to Launch")
```

```python
# Initialize data and AI assistant
users, courses, assignments, submissions, student_performance = initialize_sample_data()
ai_assistant = AIAssistant()

print("âœ… Sample data initialized!")
print(f"ğŸ“š {len(courses)} courses, {len(assignments)} assignments, {len(users)} users")
print("ğŸ¤– AI Assistant ready with all features enabled")
```

```python
# Global state and navigation functions
current_user = None
main_output = widgets.Output()

def login(user_id):
    """Login as a specific user"""
    global current_user
    current_user = users.get(user_id)
    if current_user:
        with main_output:
            clear_output()
            if current_user.role == 'teacher':
                show_teacher_dashboard()
            else:
                show_student_dashboard()

def logout():
    """Logout current user"""
    global current_user
    current_user = None
    with main_output:
        clear_output()
        show_login_screen()

def show_teacher_dashboard():
    """Display teacher dashboard"""
    render_teacher_dashboard(
        current_user, users, courses, assignments, submissions,
        ai_assistant, student_performance, widgets, HTML, plt,
        clear_output, display, show_teacher_dashboard, logout
    )

def show_student_dashboard():
    """Display student dashboard"""
    render_student_dashboard(
        current_user, users, courses, assignments, submissions,
        ai_assistant, student_performance, widgets, HTML, plt,
        defaultdict, clear_output, display, show_student_dashboard, logout, datetime
    )

def show_login_screen():
    """Display login screen"""
    render_login_screen(users, widgets, HTML, display, login)

print("âœ… Navigation system ready!")
```

```python
# Launch the LMS Application
display(HTML("""
<style>
    .widget-button {
        font-size: 14px !important;
        font-weight: 500 !important;
    }
</style>
"""))

with main_output:
    show_login_screen()

display(main_output)

print("\n" + "="*60)
print("ğŸ“ K-12 LMS PROTOTYPE - READY TO USE!")
print("="*60)
print("\nğŸ“ DEMO INSTRUCTIONS:")
print("1. Click any user button above to login")
print("2. Teacher view: Manage courses, grade assignments (try AI grading!), create new assignments")
print("3. Student view: View courses, submit assignments, see AI recommendations")
print("4. Switch between users to see different perspectives")
print("\nğŸ¤– AI FEATURES INCLUDED:")
print("   â€¢ Automated grading with intelligent feedback")
print("   â€¢ Personalized learning path recommendations")
print("   â€¢ Smart content suggestions based on performance")
print("   â€¢ Student performance prediction")
print("   â€¢ Visual analytics and progress tracking")
print("\n" + "="*60)
```
